##### 다중 분류(Multiclass Classification)
- 비대면 심리 상담사로서 메세지를 전달한 환자에 대한 감정 데이터를 수집했다.
- 각 메세지 별로 감정이 표시되어 있다.
- 미래에 동일한 메세지를 보내는 환자에게 어떤 심리 치료가 적합할 수 있는지 알아보기 위한 모델을 구축한다.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
import joblib
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
feel_df =pd.read_csv("./datasets/feeling.csv", sep=";")
feel_df
```

```python
feel_df.drop_duplicates(inplace=True)
```

```python
feel_encoder = LabelEncoder()
targets = feel_encoder.fit_transform(feel_df.feeling)
feel_df['Target'] = targets
feel_df.drop(labels=["feeling"],axis=1,inplace=True)
feel_df
```

```python
targets_0 = feel_df[feel_df.Target == 0].sample(653, random_state=124)
targets_1 = feel_df[feel_df.Target == 1].sample(653, random_state=124)
targets_2 = feel_df[feel_df.Target == 2].sample(653, random_state=124)
targets_3 = feel_df[feel_df.Target == 3]
targets_4 = feel_df[feel_df.Target == 4].sample(653, random_state=124)


feel_df = pd.concat([targets_0, targets_1,targets_2,targets_3,targets_4])
feel_df.reset_index(drop=True,inplace=True)
feel_df.Target.value_counts()
```
